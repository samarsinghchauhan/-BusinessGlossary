```python
from langchain_core.prompts import PromptTemplate
from langchain_google_genai import ChatGoogleGenerativeAI
llm_model = ChatGoogleGenerativeAI(model="gemini-2.5-flash-lite",temperature=0,max_tokens=None, api_key=GEMINI_API_KEY)
```

```python
country=""
capital=""
year=2025

question_template = """
What is then capital of {country}?
"""
weather_template = """
What is latest poulation of {capital}?
"""
```

```python
q_template = PromptTemplate(
    input_variables=["country"],
    template=question_template)

w_template = PromptTemplate(
    input_variables=["capital"],
    template=weather_template)
```

```python
Countries = ['India','United States','China']

for country in Countries:

  chain_capital = q_template | llm_model
  response = chain_capital.invoke(input = {"country":country})
  city=response.text.split("**")[1].split("**")[0]
  chain_weather = w_template | llm_model
  response = chain_weather.invoke(input = {"capital":city})
  temperature=response.text
  print(f"Capital of {country} is {city}, and tempeature today is {temperature}")

```
